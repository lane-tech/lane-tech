```
# Boas-vindas com o nome do programador

print("Bem-vindo a loja da Elane RU:4110640")

# Input do valor unitário e da quantidade do produto

valor_unitario = float(input("Digite o valor unitário do produto: R$ "))

quantidade = int(input("Digite a quantidade desejada: "))

# Cálculo do valor total sem desconto

valor_total_sem_desconto = valor_unitario * quantidade

# Aplicação de desconto conforme as condições especificadas

if valor_total_sem_desconto < 2500.00:

    desconto_percentual = 0

elif 2500.00 <= valor_total_sem_desconto < 6000.00:

    desconto_percentual = 4

elif 6000.00 <= valor_total_sem_desconto < 10000.00:

    desconto_percentual = 7

else:

    desconto_percentual = 11

# Cálculo do valor total com desconto

desconto_decimal = desconto_percentual / 100

valor_total_com_desconto = valor_total_sem_desconto * (1 - desconto_decimal)

# Exibição dos resultados

print(f"O valor sem desconto foi R$ {valor_total_sem_desconto:.2f}")

print(f"O valor com desconto foi R$ {valor_total_com_desconto:.2f} (desconto {desconto_percentual}%)")
```

```
#Boas-vindas com o nome

print("Bem-vindo a loja da Elane Amorim de Sousa")

menu = '''

--------------------Cardápio-----------------------------

------ | Tamanho |  Cupuaçu (CP)  |  Açaí (AC)   | ------

------ |    P    |     R$ 9,00    |    R$ 11,00  | ------

------ |    M    |     R$ 14,00   |    R$ 16,00  | ------

------ |    G    |     R$ 18,00   |    R$ 20,00  | ------

'''

print(menu)

# Variável para acumular o valor total do pedido

total_pedido = 0.0



#Estruturas de while, break, continue

while True:

    # Exigência de Código 2: Input e validação do sabor

    sabor = input("Escolha o sabor (cp para CUPUAÇU / ac para AÇAÍ): ")

    if sabor not in ['cp', 'ac']:

        print("Sabor inválido. Tente novamente.")

        continue  # Volta para o início do loop



    #Input e validação do tamanho

    tamanho = input("Escolha o tamanho (p/m/g): ")

    if tamanho not in ['p', 'm', 'g']:

        print("Tamanho inválido. Tente novamente.")

        continue  # Volta para o início do loop



    #Implementação de if/elif com as combinações de sabor e tamanho

    if sabor == 'cp':

        if tamanho == 'p':

            valor_pedido = 9.00

        elif tamanho == 'm':

            valor_pedido = 14.00

        else:

            valor_pedido = 18.00

        resultado = f"Cupuaçu tamanho {tamanho.upper()}"

    else:  # sabor == 'ac'

        if tamanho == 'p':

            valor_pedido = 11.00

        elif tamanho == 'm':

            valor_pedido = 16.00

        else:

            valor_pedido = 20.00

        resultado = f"Açaí tamanho {tamanho.upper()}"



    # Acumulador para somar os valores dos pedidos

    total_pedido += valor_pedido



    # Input para perguntar se deseja pedir mais alguma coisa

    continuar_pedido = input(f"Você pediu  {resultado}: R${valor_pedido:.2f}\n Deseja pedir mais alguma coisa? (s/digite outra tecla): ")

    if continuar_pedido.lower() != 's':

        break  # Sai do loop se a resposta não for 'sim'



# Mostra valor total

print('\n'"O valor total a ser pago: R${:.2f}".format(total_pedido))
```

```
# Função para escolher o serviço

def escolha_servico():

    while True:

        servico = input("Entre com o tipo de serviço desejado\nDIG - Digitalização\nICO - Impressão Colorida\nIBO - Impressão Preto e Branco\nFOT - Fotocópia\n").upper()

        if servico in ["DIG", "ICO", "IBO", "FOT"]:

            if servico == "DIG":

                return 1.10

            elif servico == "ICO":

                return 1.00

            elif servico == "IBO":

                return 0.40

            elif servico == "FOT":

                return 0.20

        else:

            print("Escolha Inválida'\n. Entre com o tipo de serviço desejado novamente.'\n")



# Função para obter o número de páginas com desconto

def num_paginas_com_desconto():

    while True:

        try:

            num_paginas = int(input("\nEntre com o número de páginas:'\n"))

            if num_paginas < 20:

                return num_paginas

            elif 20 <= num_paginas < 200:

                return num_paginas * 0.85

            elif 200 <= num_paginas < 2000:

                return num_paginas * 0.80

            elif 2000 <= num_paginas < 20000:

                return num_paginas * 0.75

            else:

                print("Não aceitamos tantas páginas de uma vez.\nPor favor, entre com o número de páginas novamente.")

        except ValueError:

            print("Digite um valor numérico válido. Tente novamente.")



# Função para escolher serviço adicional

def servico_extra():

    while True:

        try:

            adicional = int(input("Deseja adicionar mais algum serviço?'\n 1 - Encadernação Simples\n 2 - Encadernação Capa Dura\n 0 - Não desejo mais nada\n"))

            if adicional in [1, 2, 0]:

                return adicional

            else:

                print("Escolha Inválida'\n. Entre com o tipo de serviço desejado novamente.")

        except ValueError:

            print("Digite um valor numérico válido. Tente novamente.")



# Função principal

def main():

    try:

        # Mensagem de boas-vindas

        print("Bem-vindo ao sistema de cobrança da copiadora da Elane RU:4110640")



        # Escolha do serviço

        servico = escolha_servico()



        # Número de páginas com desconto

        num_paginas_desconto = num_paginas_com_desconto()



        # Serviço adicional

        adicional = servico_extra()



        # Cálculo do total a pagar

        total_pagar = servico * num_paginas_desconto



        # Adicional

        if adicional == 1:

            total_pagar += 15

        elif adicional == 2:

            total_pagar += 40



        # Exibição do total a pagar

        print(f"Total (R$): {total_pagar:.2f} (serviço {servico} * páginas: {num_paginas_desconto} + extras: {15 if adicional == 1 else 40 if adicional == 2 else 0})")



    except Exception as e:

        print(f"Erro: {e}")



if __name__ == "__main__":

    main()
```
